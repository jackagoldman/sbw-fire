```python
import networkx as nx
from matplotlib import pyplot as plt
import graphviz # pygraphviz should be available
import pydot
```

```python
dot1 = graphviz.Digraph('chapter3-step1-dag', comment='step 1 dag') 
dot1.node('T', 'Time Since Defoliation', shape = "rect")  
dot1.node('D', 'Defoliated', shape = "rect")
dot1.node('Y', 'Number of Years Defoliated', shape = "rect")
dot1.node('C', 'Defoliation Intensity',shape = "rect")
dot1.node('B', 'Burn Severity',  shape = "rect", color = "red")
dot1.node('R', 'Forest Recovery', shape = "ellipse", color="red")


dot1.edges(['TC', 'DC', 'YC', 'CB', 'BR', 'CR'])
#dot2.edge('B', 'R', style= "dashed")
#dot2.edge('S', 'R', style= "dashed")



dot1
```
